QUICK

```python
# Quicksort tiene una complejidad promedio de O(n log n)
# Sin embargo, en el peor de los casos puede llegar a O(n^2) si la elección del pivote no es óptima

def quick(arr):
    # Verificar si el tamaño del arreglo es menor o igual a 1
    if len(arr) <= 1:
        return arr

    # Seleccionar un pivote
    pivot = arr[len(arr) // 2]

    # Dividir el arreglo en tres partes: menores, iguales y mayores que el pivote
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    # Combinar recursivamente las sublistas ordenadas
    return quick(left) + middle + quick(right)

# Obtener los datos de la URL
ruta = "https://www.datos.gov.co/resource/u37r-hjmu.json"
respuesta = requests.get(ruta)
datos = json.loads(respuesta.text)

# Convertir los datos JSON en un DataFrame de pandas
df = pd.DataFrame(datos)

# Eliminar filas con valores NaN en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Convertir la columna 'mod_comuni_escrita_punt' a números enteros, omitiendo los valores no numéricos
df['mod_comuni_escrita_punt'] = pd.to_numeric(df['mod_comuni_escrita_punt'], errors='coerce')

# Eliminar filas con valores no numéricos en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Obtener los datos ordenados utilizando Quicksort

datos_ordenados = quick(df['mod_comuni_escrita_punt'].tolist())

# Mostrar los datos ordenados
print(datos_ordenados)
```

COUNTING

```python
def countingsort(arr):
    if len(arr) == 0:
        return arr

    # Encontrar el rango de los elementos
    min_val = min(arr)  # O(n)
    max_val = max(arr)  # O(n)
    range_val = max_val - min_val + 1  # O(1)

    # Inicializar el arreglo de conteo y el arreglo de salida
    count = [0] * range_val  # O(k)
    output = [0] * len(arr)  # O(n)

    # Contar la frecuencia de cada elemento
    for num in arr:  # O(n)
        count[num - min_val] += 1

    # Calcular las posiciones finales de cada elemento
    for i in range(1, range_val):  # O(k)
        count[i] += count[i - 1]

    # Colocar los elementos en el arreglo de salida
    for i in range(len(arr) - 1, -1, -1):  # O(n)
        output[count[arr[i] - min_val] - 1] = arr[i]
        count[arr[i] - min_val] -= 1

    return output

# Obtener los datos de la URL
ruta = "https://www.datos.gov.co/resource/u37r-hjmu.json"
respuesta = requests.get(ruta)
datos = json.loads(respuesta.text)

# Convertir los datos JSON en un DataFrame de pandas
df = pd.DataFrame(datos)

# Eliminar filas con valores NaN en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Convertir la columna 'mod_comuni_escrita_punt' a números enteros, omitiendo los valores no numéricos
df['mod_comuni_escrita_punt'] = pd.to_numeric(df['mod_comuni_escrita_punt'], errors='coerce')

# Eliminar filas con valores no numéricos en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Obtener los datos ordenados utilizando Counting Sort
# El tiempo de ejecución del Counting Sort es O(n + k), donde "n" es el número de elementos en el arreglo
# y "k" es el rango de los elementos.
datos_ordenados = countingsort(df['mod_comuni_escrita_punt'].astype(int).tolist())

# Mostrar los datos ordenados
print(datos_ordenados)
```

Bucket

```python


def bucket_sort(arr):
    # Número de buckets
    num_buckets = 10

    # Crear buckets vacíos
    buckets = [[] for _ in range(num_buckets)]

    # Distribuir los elementos en los buckets
    for num in arr:
        index = min(num // 10, num_buckets - 1)  # Calcular el índice del bucket
        buckets[index].append(num)

    # Ordenar cada bucket
    for i in range(num_buckets):
        buckets[i].sort()  # El tiempo de ordenamiento de cada bucket es O(m log m), donde "m" es el número de elementos en el bucket

    # Combinar los buckets ordenados
    sorted_arr = []
    for bucket in buckets:
        sorted_arr.extend(bucket)

    return sorted_arr

# Obtener los datos de la URL
ruta = "https://www.datos.gov.co/resource/u37r-hjmu.json"
respuesta = requests.get(ruta)
datos = json.loads(respuesta.text)

# Convertir los datos JSON en un DataFrame de pandas
df = pd.DataFrame(datos)

# Eliminar filas con valores NaN en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Convertir la columna 'mod_comuni_escrita_punt' a números enteros, omitiendo los valores no numéricos
df['mod_comuni_escrita_punt'] = pd.to_numeric(df['mod_comuni_escrita_punt'], errors='coerce')

# Eliminar filas con valores no numéricos en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Obtener los datos ordenados utilizando Bucket Sort
# La complejidad temporal del Bucket Sort es O(n log n) en el caso promedio.
datos_ordenados = bucket_sort(df['mod_comuni_escrita_punt'].astype(int).tolist())

# Mostrar los datos ordenados
print(datos_ordenados)
```

MERGE PAI

```python


def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Dividir el arreglo en dos mitades
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    # Aplicar merge_sort recursivamente a ambas mitades
    left = merge_sort(left)
    right = merge_sort(right)

    # Combinar ambas mitades ordenadas
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    # Combinar los elementos de ambas mitades en orden
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Agregar los elementos restantes de la mitad izquierda
    result.extend(left[i:])

    # Agregar los elementos restantes de la mitad derecha
    result.extend(right[j:])

    return result

# Obtener los datos de la URL
ruta = "https://www.datos.gov.co/resource/u37r-hjmu.json"
respuesta = requests.get(ruta)
datos = json.loads(respuesta.text)

# Convertir los datos JSON en un DataFrame de pandas
df = pd.DataFrame(datos)

# Eliminar filas con valores NaN en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Convertir la columna 'mod_comuni_escrita_punt' a números enteros, omitiendo los valores no numéricos
df['mod_comuni_escrita_punt'] = pd.to_numeric(df['mod_comuni_escrita_punt'], errors='coerce')

# Eliminar filas con valores no numéricos en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Obtener los datos ordenados utilizando Merge Sort
# La complejidad temporal del Merge Sort es O(n log n) en todos los casos.
datos_ordenados = merge_sort(df['mod_comuni_escrita_punt'].astype(int).tolist())

# Mostrar los datos ordenados
print(datos_ordenados)
```

```python

# Función para ordenar una lista utilizando el algoritmo de Insertion Sort
def insertion_sort(arr):
    # Iterar a través de la lista, comenzando desde el segundo elemento
    for i in range(1, len(arr)):
        # Guardar el valor actual para comparación
        key = arr[i]
        # Inicializar un índice para comparar con el elemento anterior
        j = i - 1
        # Mover los elementos mayores que el valor actual una posición hacia adelante
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        # Colocar el valor actual en su posición correcta en la lista ordenada
        arr[j + 1] = key

# Obtener los datos de la URL
ruta = "https://www.datos.gov.co/resource/u37r-hjmu.json"
respuesta = requests.get(ruta)
datos = json.loads(respuesta.text)

# Convertir los datos JSON en un DataFrame de pandas
df = pd.DataFrame(datos)

# Eliminar filas con valores NaN en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Convertir la columna 'mod_comuni_escrita_punt' a números enteros, omitiendo los valores no numéricos
df['mod_comuni_escrita_punt'] = pd.to_numeric(df['mod_comuni_escrita_punt'], errors='coerce')

# Eliminar filas con valores no numéricos en la columna 'mod_comuni_escrita_punt'
df = df.dropna(subset=['mod_comuni_escrita_punt'])

# Obtener los datos ordenados utilizando Insertion Sort
# La complejidad temporal del Insertion Sort es O(n^2) en el peor caso.
datos_ordenados = df['mod_comuni_escrita_punt'].astype(int).tolist()
insertion_sort(datos_ordenados)

# Mostrar los datos ordenados
print(datos_ordenados)
```
